```python
neuron = pd.read_csv("neuron.csv")
neuron
```

```python
import matplotlib.pyplot as plt


fig, axes = plt.subplots(1, 3, figsize=(20, 6)) # Create a figure and a set of subplots

membrane_potential_column = 'Membrane potential (mV)'
firing_rate_column = 'Membrane potential (mV)'
stimulus_column = 'Stimulus'

# Histogram for the whole data
axes[0].hist(neuron[firing_rate_column], bins=20, edgecolor='black')
axes[0].set_title('Membrane Potential Distribution (Whole Data)')
axes[0].set_xlabel('Membrane Potential (mV)')
axes[0].set_ylabel('Frequency')

# Histogram when stimulus is on
# Filter for stimulus == 1
axes[1].hist(neuron[neuron[stimulus_column] == 1][firing_rate_column], bins=20, edgecolor='black', color='orange')
axes[1].set_title('Membrane Potential Distribution (Stimulus ON)')
axes[1].set_xlabel('Membrane Potential (mV)')
axes[1].set_ylabel('Frequency')

# Histogram when stimulus is off
# Filter for stimulus == 0
axes[2].hist(neuron[neuron[stimulus_column] == 0][firing_rate_column], bins=20, edgecolor='black', color='green')
axes[2].set_title('Membrane Potential Distribution (Stimulus OFF)')
axes[2].set_xlabel('Membrane Potential (mV)')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
